```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
SAMPLES_PER_SECOND = 100_000

df = pd.read_csv('sys_with_captouch_1V8.csv')
df.columns
```

```python
# Downsample the dataframe by a factor of 10
df_downsampled = df.iloc[::10, :]

# Convert Timestamp from milliseconds to seconds
df_downsampled['Time(s)'] = df_downsampled['Timestamp(ms)'] / 1000

# Plot the downsampled dataframe
plt.figure(figsize=(10, 6))
plt.plot(df_downsampled['Time(s)'], df_downsampled['Current(uA)'])
plt.xlabel('Time [s]')
plt.ylabel('Current [µA]')
plt.title('Current measurement of Bluetooth system with touch sensor')
plt.show()
```
