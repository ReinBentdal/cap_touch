```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constants
C_fixed = 1.0  # Fixed capacitor value in microfarads (µF)
C_variable = np.linspace(0.1, 50, 500)  # Variable capacitor values in µF

# Total capacitance in series
C_total = 1 / (1 / C_fixed + 1 / C_variable)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(C_variable, C_total, label=r'$C_{\mathrm{total}}$', color='blue')
plt.axhline(C_fixed, color='red', linestyle='--', label=r'$C_{\mathrm{fixed}} = 1 \, \mu\mathrm{F}$')
plt.axvline(1, color='green', linestyle='--', label=r'$C_{\mathrm{variable}} = 1 \, \mu\mathrm{F}$')

# Labels and title
plt.title('Total Capacitance of Two Capacitors in Series', fontsize=14)
plt.xlabel(r'$C_{\mathrm{variable}}$ ($\mu$F)', fontsize=12)
plt.ylabel(r'$C_{\mathrm{total}}$ ($\mu$F)', fontsize=12)
plt.grid(alpha=0.5)
plt.legend()
```
